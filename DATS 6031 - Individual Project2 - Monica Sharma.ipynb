```python
# Best Seller books on Paytm

import math
import pandas as pd
import matplotlib.pyplot as pp
import plotly
from plotly import tools
import plotly.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

py.sign_in('rahulsethi80','g5jVVUkimt1N2LHA4vvo')
plotly.offline.init_notebook_mode()
```

```python
#Reading the data from csv file 
df_data=pd.read_csv("paytm_com-ecommerce_sample.csv")
df_data.head(4)
```

```python
#Dropping the columns  Country and Type as the values are same throughout (Country=IN; Type=Bestseller)

df_data=df_data.drop(['Country','Type'],axis=1)
```

```python
#Dimension of Data
df_data.shape
```

```python
# Removing na Values
df_data.dropna(how='any',inplace=True)
df_data.shape
```

```python
#Dataset after removing na values
df_data.head(3)
```

```python
#The Genre column includes all the different types of gnere that the book belongs and is separted by '|'. 
#It also includes the book name as the last field in it. Hence creating a new data frame to include genre values without book name

df_genre_without_bookname = df_data
df_genre_without_bookname.head(3)
```

```python
#Splitting the Genre column based on '|'

#Split the column Genre to remove the bookname which occurs after the last delimeter '|'
genre_book = df_genre_without_bookname.Genre.str.split('|').str[:-1].str.join('|')
genre_book.head()

#Split the column Genre into different rows based on delimeter '|'
genre_book = genre_book.str.split('|', expand=True).stack().str.strip().reset_index(level=1, drop=True)
```

```python
#Drop the column Genre from the data frame df_genre_without_bookname
df_genre_without_bookname = df_genre_without_bookname.drop('Genre', axis=1)
```

```python
#Adding the new Genre column to the data frame df_genre_without_bookname

df_genre_without_bookname = pd.concat([df_genre_without_bookname,genre_book],axis=1).reset_index(drop=True)
df_genre_without_bookname.head(5)
```

```python
#Renaming the newly added column as Genre

df_genre_without_bookname = df_genre_without_bookname.rename(columns = {0:'Genre'})
df_genre_without_bookname.head(2)
```

```python
#Dimension of the data frame df_genre_without_bookname
df_genre_without_bookname.shape
```

```python
#Rearranging the columns to get Genre at the first position

cols = df_genre_without_bookname.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_genre_without_bookname = df_genre_without_bookname[cols]
df_genre_without_bookname.head(3)
```

```python
#Counting the number of books in each Genre

genre_data = pd.melt(df_genre_without_bookname, value_vars=['Genre'], value_name='Book Genres')
genre_count = genre_data.groupby(by=['Book Genres'])['Book Genres'].count()
genre_count.head(15)
```

```python
#Plotting a Donut Pie chart to display the number of books in each Genre

labels1 = df_genre_without_bookname['Genre']
values1 = genre_count

trace1 = go.Pie(labels=labels1, values=values1,pull=.2,hole=.2)

py.iplot([trace1], filename='basic_pie_chart', legend = True)
```

```python
#Top 10 Costliest book sold on paytm 

#Sort based on Selling_price
sort_selling_price = df_data.sort_values(by='Selling_price', ascending= False) 
sort_selling_price = sort_selling_price.head(10)
sort_selling_price
```

```python
#Plotting graph to display the top 10 costliest book sold by Paytm
trace2 = go.Histogram2d(
    x=sort_selling_price['Name'],
    y=sort_selling_price['Selling_price'])
    
fig = tools.make_subplots(rows=1, cols=1)
fig.append_trace(trace2, 1, 1)

fig['layout'].update(height=600, width=1000, title='Top 10 Costliest books sold by paytm website')
py.iplot(fig)
```

```python
#Sorting by brands from 1936 to 2016
indexed_df_data = df_data.set_index(['Name'])
top_brand= indexed_df_data.sort_values(by='Brand', ascending= False) 
top_brand = top_brand.head(4)
top_brand
```

```python
#Creating a new column Year by Extracting it from column Specification 

df_data['Year'] = df_data['Specifications'].str.split('\"publication_year\"=>\"').str[1].str[:4]
df_data.head(2)

#Fetch 2015 Brands

data_year_indexed = df_data.set_index(['Year','Name'])
data_year_indexed_2015 = data_year_indexed.loc['2015']
brands_2015 = data_year_indexed_2015.sort_values(by='Brand', ascending=False).head(100)

#Fetch 2016 Brands

data_year_indexed_2016 = data_year_indexed.loc['2016']
brands_2016 = data_year_indexed_2016.sort_values(by='Brand', ascending=False).head(100)
```

```python
# Plot graph to compare the brands of the book from the year 2015 and 2016 

trace1 = go.Histogram(
    x = brands_2015['Brand'],
    name = '2015'
)
trace2 = go.Histogram(
    x = brands_2016['Brand'],
    name = '2016'
)

data2 = [trace1, trace2]
layout1 = go.Layout(
    title='Comparision of books brands between the year 2015 and 2016',
    xaxis=dict(
        title="Brand names of books"
    ),
    yaxis=dict(
        title="Count of books"
    ),
    
)

fig = go.Figure(data=data2, layout=layout1),
py.iplot(data2, filename='histogram')
```

```python
#Calculate the total earnings by Paytm yearwise from 2012 to 2016.

data_year_indexed_2012 = data_year_indexed.loc['2012']
sum2012 = sum(data_year_indexed_2012['Selling_price'])

data_year_indexed_2013 = data_year_indexed.loc['2013']
sum2013 = sum(data_year_indexed_2013['Selling_price'])

data_year_indexed_2014 = data_year_indexed.loc['2014']
sum2014 = sum(data_year_indexed_2014['Selling_price'])

sum2015=sum(data_year_indexed_2015['Selling_price'])
sum2016=sum(data_year_indexed_2016['Selling_price'])

```

```python
#Making a line plot to display the total earnings by Paytm from 2012 to 2016.

trace=go.Scatter(
    x=['2012','2013','2014','2015','2016'],
    y=[sum2012,sum2013,sum2014,sum2015,sum2016],
    marker=dict(size=20)
)

fig = tools.make_subplots(rows=1, cols=1)
fig.append_trace(trace, 1, 1)

fig['layout'].update(height=600, width=1000, title='Earnings made by Paytm for the year 2012 to 2016')
py.iplot(fig)
```

```python
#Calculating the top 10  discount offered by paytm on books for the year 2013 and 2016

data_year_indexed = df_data.set_index(['Year'])

top_discount_2013 = data_year_indexed.loc['2013'].sort_values(by='Discount', ascending= False).head(10)
top_discount_2016 = data_year_indexed.loc['2016'].sort_values(by='Discount', ascending= False).head(10)
```

```python
#Plot an area plot to see top 10 discount offered by paytm on books for the year 2013 and 2016


trace1 = go.Scatter(
    x=top_discount_2013['Name'],
    y=top_discount_2013['Discount'],
    fill='tonexty',
    name='2013'
)
trace2 = go.Scatter(
    x=top_discount_2016['Name'],
    y=top_discount_2016['Discount'],
    fill='tonexty',
    name='2016'
)

fig = tools.make_subplots(rows=1, cols=2)

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)

fig['layout'].update(height=600, width=1000, title='Maximum discounts given by paytm for the year 2013 and 2016')
py.iplot(fig)
```
